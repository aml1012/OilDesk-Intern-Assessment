### Question 1: Python Basics and Data Manipulation
**Objective:** Demonstrate basic Python skills and data manipulation using Pandas.

**Task:**
- Load a given CSV file containing metal prices into a Pandas DataFrame. DONE
- Filter the data to include only 'Copper' and 'Zinc' for the year 2021. DONE
- Calculate the average price per month for each metal and plot it. DONE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## checking the data
data = pd.read_csv(r"C:\Users\lagan\Desktop\OilDesk-Intern-Assessment\data\MarketData.csv")
data.columns = ['Date','LME_Copper_3MO', 'LME_Aluminum_3MO', 'LME_Zinc_3MO','LME_Lead_3MO', 'LME_Tin_3MO', 'Generic_CL_Future']
data.head(20)
```

```python
# removing junk rows
data = data.drop(index = data.index[:6])
data.head(20)
```

```python
# checking datatypes
print(data.dtypes)
```

```python
# set date as the index 
data['Date']= pd.to_datetime(data['Date'], dayfirst = True)
data.set_index('Date', inplace=True)
```

```python
# convert prices to numeric data types
for metal in data.columns:
    data[metal] = pd.to_numeric(data[metal])
    
data.dtypes
```

```python
# filtering data
filtered_data = data[['LME_Copper_3MO', 'LME_Zinc_3MO']]
filtered_data.head(20)
```

```python
#calculate monthly avg prices and store in a df
monthly_avg_prices = filtered_data.groupby(filtered_data.index.to_period('M')).mean()
monthly_avg_prices.head(20)
```

```python
## plot monthly prices
monthly_avg_prices.index = monthly_avg_prices.index.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_prices.index, monthly_avg_prices['LME_Copper_3MO'], label='LME Copper 3MO', color='orange')
plt.plot(monthly_avg_prices.index, monthly_avg_prices['LME_Zinc_3MO'], label='LME Zinc 3MO', color='blue')
plt.xlabel('Date')
plt.ylabel('Average Last Settlement Price ($)')
plt.title('Monthly Average Last Settlement Price for Copper and Zinc')

plt.legend()
plt.grid(True)
plt.show()

```
